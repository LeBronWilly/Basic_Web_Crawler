```python
# 需求：目標網站是PTT寶可夢版，我們想要爬取那一頁所有文章標題的文字字串，並將每個字串一個個存放在list中
# Hint：前置作業不可少！使用.select方法會比較好做！✬此題需要使用上午場所教的for迴圈搭配list資料型態的處理







```

## 解答

```python
# 前置作業(完全與Preparation一樣)
import requests
from bs4 import BeautifulSoup
ptt_html = requests.get("https://www.ptt.cc/bbs/PokemonGO/index.html")
pttbs4 = BeautifulSoup(ptt_html.text, 'html.parser')
```

```python
pttbs4.select(".title")
```

```python
PTT_article_titles = []  # 新創一個空List，用來存放所抓取的標題

# 把文字萃取出來，因為我們有多個資料存放在這個pttbs4.select(".title a")的List中，所以需要使用for迴圈將逐筆將List中的資料一個個輸出
for i in pttbs4.select(".title a"):
    print(i.string)  #  使用print呈現給我們看
    PTT_article_titles.append(i.string)  #  使用.string把非文字的部分去除掉，並且使用.append方法一個一個加入list裡面
```

# **Application 2✬✬**
**爬蟲實作2－PTT板多頁面標題爬取**

```python
# -*- coding: utf-8 -*-
"""
Created on Wed July 08 23:16:04 2020
@author: Willy Fang

"""

# 前置作業
import requests, time  # 另外導入time時間套件/模組
from bs4 import BeautifulSoup # 導入BeautifulSoup模組：解析HTML語法工具
```

```python
# 創立一個爬蟲函數
def get_PTT_onepage(URL):  # 參數是一個字串->網址
    # 設定Header與Cookie
    my_headers = {'cookie': 'over18=1;'}
    # 發送get請求到PPT寶可夢版
    ptt_html = requests.get(URL, headers = my_headers)
    # 把網頁程式碼(HTML) 丟入 BeautifulSoup解析
    pttbs4 = BeautifulSoup(ptt_html.text, "html.parser")
    # 萃取文字出來。
    # 因為我們有多個Tags存放在 List 中，所以需要使用for 迴圈將逐筆取出
    for t in pttbs4.select(".title a"):  # 如同Application 1所做的
        print(t.string)
        PTT_multipages_titles.append(t.string)
```

```python
start = 750 # 設定起始網頁 (可自行調整)
number = 10    # 設定要從開始頁面，往後或往前爬多少頁(可自行調整)
# end = start - number   # 頁數往前爬的作法
end = start + number  # 頁數往後爬的作法
```

```python
PTT_multipages_titles = []  # 新創一個空List，用來存放所抓取的標題

# for i in range(start,end,-1):  # 頁數往前爬的作法
for i in range(start,end,1):  # 頁數往後爬的作法
    # 組合URL
    link = "https://www.ptt.cc/bbs/PokemonGO/index" + str(i) + ".html"
    # 執行單頁面網頁爬蟲
    print("==============目前在爬取第",i,"頁......==============")
    get_PTT_onepage(link)
    # 避免request太快，被PTT認為是DDOS攻擊而被封鎖請求
    time.sleep(1) # 間隔1秒
```

```python
PTT_multipages_titles
```

# **Application 3✬✬✬**
**爬蟲實作3－PTT板文章內文爬取**

```python
# -*- coding: utf-8 -*-
"""
Created on Fri July 10 22:37:52 2020
@author: Willy Fang

"""
# 前置作業
import requests 
from bs4 import BeautifulSoup

# 文章連結
# URL = "https://www.ptt.cc/bbs/Gossiping/M.1590853964.A.10B.html"  # 八卦版的某篇文章(這個就有未滿18的問題了)
# URL = "https://www.ptt.cc/bbs/Kaohsiung/M.1590844585.A.A1C.html"  # 高雄版的某篇文章
URL = "https://www.ptt.cc/bbs/PokemonGO/M.1590132148.A.511.html"  # 寶可夢版的某篇文章


my_headers = {'cookie': 'over18=1;'}
ptt_html = requests.get(URL, headers = my_headers)
pttbs4 = BeautifulSoup(ptt_html.text,"html.parser")
```

```python
# PTT文章的上方4個欄位（作者、看板、title標題、日期）
header = pttbs4.find_all('span','article-meta-value')  # 取得"所有"符合條件的<span>標籤資料內容，其中屬性需含有class="article-meta-value"
# header = pttbs4.select("span.article-meta-value")  # 這裡若有空格如pttbs4.select("span .article-meta-value")則會回傳空值
header # 查看header資料長怎樣
```

```python
# 作者
author = header[0].text # 這裡用.string也可以！
# 看版
board = header[1].text
# 標題
title = header[2].text
# 日期
date = header[3].text


# 查找所有html元素，抓取內容，通常會抓有id屬性的
# main_container = pttbs4.find(id='main-container') 
main_container = pttbs4.find_all(id='main-container')[0]  # 由於id比較有獨一性，所以兩者都可以

all_text = main_container.text  # 把所有文字都抓取出來(必須要用.text，因為要包含到子標籤的內容)
all_text # 查看all_text資料長怎樣......一團亂，因此我們需要直接從瀏覽器網頁來看
```

```python
pre_text = all_text.split('※ 發信站')[0]  # 把整個內容切割透過 "※ 發信站" 切割成2個陣列，然後取前面那一個
texts = pre_text.split('\n')  # 把每段文字，根據'\n'切開
texts # 查看texts資料長怎樣
```

```python
contents = texts[2:] # 若單看texts你會發現，PTT文章的上方4個欄位（作者、看板、title標題、日期）也都包含在裡頭，因此就透過slice方式把他們篩掉
# 內容
content = '\n'.join(contents) # 使用.join的方法，透過分段的特殊字元('\n')，把list裡面的所有元素串接起來成一整個字串(str)


# 結果顯示
print('作者：'+author)
print('看板：'+board)
print('標題：'+title)
print('日期：'+date)
print('內容：'+content)
```
